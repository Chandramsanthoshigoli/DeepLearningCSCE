Fitting the Linear Regression Model

```python
import statsmodels.api as sm
numeric_data = data_swarna.select_dtypes(include=[np.number])
X = numeric_data.drop(columns=['Yearly Amount Spent'])  
y = numeric_data['Yearly Amount Spent'] 
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

Recommendation:

OLS regression analysis reveals a highly effective model for predicting Yearly Amount Spent, with an impressive R-squared value of 0.984. Three key factors significantly influence spending: Average Session Length, Time on App, and Length of Membership, with coefficients of $25.73, $38.71, and $61.58 respectively, all showing strong positive relationships with yearly expenditure. Interestingly, Time on Website appears to have minimal impact, lacking statistical significance. The model's overall validity is supported by a high F-statistic of 7766 and a p-value of 0.00. However, a large condition number (2.64e+03) suggests potential multicollinearity, warranting caution in interpreting individual coefficients. Despite this limitation, the analysis strongly indicates that strategies focusing on extending user engagement, particularly through the app and encouraging longer memberships, could effectively boost customer spending. The model's high explanatory power provides valuable insights for targeted marketing and product development, though further investigation into the role of website interaction may be beneficial.
